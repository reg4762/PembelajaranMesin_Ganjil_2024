```python
import os
import cv2
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Path to image directories
train_dir = "/content/drive/MyDrive/Pembelajaran Mesin/week4/data/images/images/training"
test_dir = "/content/drive/MyDrive/Pembelajaran Mesin/week4/data/images/images/test"
```

```python
# Fungsi untuk mengonversi gambar menjadi fitur histogram
def extract_histogram(image, bins=(8, 8, 8)):
    # Mengonversi gambar ke ruang warna HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    # Menghitung histogram pada setiap channel HSV
    hist = cv2.calcHist([hsv_image], [0, 1, 2], None, bins, [0, 180, 0, 256, 0, 256])
    # Menormalkan histogram
    hist = cv2.normalize(hist, hist).flatten()
    return hist
```

```python
# Fungsi untuk memuat gambar dan label dari direktori
def load_images_from_directory(directory):
    images = []
    labels = []

    for subdir, dirs, files in os.walk(directory):
        for file in files:
            # Memuat setiap gambar
            img_path = os.path.join(subdir, file)
            image = cv2.imread(img_path)

            # Mengekstrak fitur histogram
            hist = extract_histogram(image)
            images.append(hist)

            # Label berdasarkan nama folder
            label = subdir.split('/')[-1]
            labels.append(label)

    return np.array(images), np.array(labels)
```

```python
# Memuat data training
X_train, y_train = load_images_from_directory(train_dir)
# Memuat data testing
X_test, y_test = load_images_from_directory(test_dir)
```

```python
# Membuat model SVM dengan kernel linear
model = SVC(kernel='linear')

# Melatih model dengan data training
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

# Memprediksi label pada data testing
y_pred = model.predict(X_test)

# Menghitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print(f'Akurasi Model: {accuracy * 100:.2f}%')

# Menampilkan laporan klasifikasi
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
